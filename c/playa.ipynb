```python
stats = player_d1[['overall','age']].values
type_label = np.where(player_d1['over50M']=='y', 0, 1)

# Feature names
stat_features = player_d1.columns.values[1:].tolist()
stat_features
```

```python
model = svm.SVC(kernel='linear')
model.fit(stats, type_label)
```

```python
w = model.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(80,100)
yy = np.linspace(20,35) 
b = model.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = model.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])
```

```python
sns.lmplot('overall', 'age', data=player_d1, hue='over50M', palette='Set1', fit_reg=False, scatter_kws={"s": 40})
plt.plot(xx,yy,linewidth=2,color='black')
plt.scatter(model.support_vector_[:,50],model.support_vectors_[:,1],s=80,facecolors='node')
```
